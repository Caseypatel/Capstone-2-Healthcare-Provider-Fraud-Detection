```python
#Replacing 2 with 0 for chronic conditions ,that means chroniv condition No is 0 and yes is 1

train_ben = train_ben.replace({'ChronicCond_Alzheimer': 2, 'ChronicCond_Heartfailure': 2,
                                                       'ChronicCond_KidneyDisease': 2, 'ChronicCond_Cancer': 2, 
                                                       'ChronicCond_ObstrPulmonary': 2, 'ChronicCond_Depression': 2, 
                                                       'ChronicCond_Diabetes': 2, 'ChronicCond_IschemicHeart': 2, 
                                                       'ChronicCond_Osteoporasis': 2, 'ChronicCond_rheumatoidarthritis': 2,
                                                       'ChronicCond_stroke': 2 }, 0)

train_ben = train_ben.replace({'RenalDiseaseIndicator': 'Y'}, 1)

test_ben = test_ben.replace({'ChronicCond_Alzheimer': 2, 'ChronicCond_Heartfailure': 2, 
                                                     'ChronicCond_KidneyDisease': 2, 'ChronicCond_Cancer': 2, 
                                                     'ChronicCond_ObstrPulmonary': 2, 'ChronicCond_Depression': 2, 
                                                     'ChronicCond_Diabetes': 2, 'ChronicCond_IschemicHeart': 2, 
                                                     'ChronicCond_Osteoporasis': 2, 'ChronicCond_rheumatoidarthritis': 2, 
                                                     'ChronicCond_stroke': 2 }, 0)

test_ben = test_ben.replace({'RenalDiseaseIndicator': 'Y'}, 1)
```

```python
test_ben.head()
train_ben.head()
```

```python
## Lets Create Age column to the dataset

train_ben['DOB'] = pd.to_datetime(train_ben['DOB'] , format = '%Y-%m-%d')
train_ben['DOD'] = pd.to_datetime(train_ben['DOD'],format = '%Y-%m-%d',errors='ignore')
train_ben['Age'] = round(((train_ben['DOD'] - train_ben['DOB']).dt.days)/365)


test_ben['DOB'] = pd.to_datetime(test_ben['DOB'] , format = '%Y-%m-%d')
test_ben['DOD'] = pd.to_datetime(test_ben['DOD'],format = '%Y-%m-%d',errors='ignore')
test_ben['Age'] = round(((test_ben['DOD'] - test_ben['DOB']).dt.days)/365)
train_ben.head()
```

```python
## As we see that last DOD value is 2009-12-01 ,which means Beneficiary Details data is of year 2009.
## so we will calculate age of other benficiaries for year 2009.

train_ben.Age.fillna(round(((pd.to_datetime('2009-12-01' , format = '%Y-%m-%d') 
                                         - train_ben['DOB']).dt.days)/365), inplace=True)


test_ben.Age.fillna(round(((pd.to_datetime('2009-12-01' , format = '%Y-%m-%d') 
                                        - test_ben['DOB']).dt.days)/365), inplace=True)

# converting age from flot to int
train_ben['Age'] = train_ben['Age'].astype(int)
test_ben['Age'] = test_ben['Age'].astype(int)
```

```python
train_ben.info()
train_ben.head()
```

```python
#Lets create a new variable 'WhetherDead' with flag 1 means Dead and 0 means not Dead

train_ben.loc[train_ben.DOD.isna(),'WhetherDead']=0
train_ben.loc[train_ben.DOD.notna(),'WhetherDead']=1
train_ben.loc[:,'WhetherDead'].head(7)


test_ben.loc[test_ben.DOD.isna(),'WhetherDead']=0
test_ben.loc[test_ben.DOD.notna(),'WhetherDead']=1
test_ben.loc[:,'WhetherDead'].head(3)

train_ben.head()
```

```python
## As patient can be admitted for only for 1 day,we will add 1 to the difference of Discharge Date and Admission Date 

test_ip['AdmissionDt'] = pd.to_datetime(test_ip['AdmissionDt'] , format = '%Y-%m-%d')
test_ip['DischargeDt'] = pd.to_datetime(test_ip['DischargeDt'],format = '%Y-%m-%d')
test_ip['AdmitForDays'] = ((test_ip['DischargeDt'] - test_ip['AdmissionDt']).dt.days)+1

test_ip.head()
```
